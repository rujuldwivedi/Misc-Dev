```python
import random
from matplotlib import pyplot as plt

p = 0.5
n = 10000

str_len = []

for i in range(n):

  l = 0

  while (random.uniform(0,1) < p):
    pass
  
  while (random.uniform(0,1) > p):
    l += 1

  l += 1

  str_len += [l]


str_set = set(str_len)


str_dict = {}
for i in str_set:
  str_dict[i] = 0

for i in str_len:
  str_dict[i] += 1/n


x = list(str_dict.keys())
y = list(str_dict.values())

plt.bar(x,y,color='blue',edgecolor='red',width=0.8)
plt.title("Empirical Distribution")
plt.xlabel("Length of String")
plt.ylabel("Probability")
plt.show()

k_list = []

for i in range(n):
  k = 0
  while (random.uniform(0,1) > p):
    k += 1
  k += 1
  k_list += [k]

k_set = set(k_list)
k_prob = {}

for i in k_set:
  k_prob[i] = 0

for i in k_list:
  k_prob[i] += 1/n

x = list(k_prob.keys())
y = list(k_prob.values())

plt.bar(x,y,color='blue',edgecolor='red',width=0.8)
plt.title("Theoretical : Geometric Distribution")
plt.xlabel("No of Successes")
plt.ylabel("Probability")
plt.show()



```
