```python
import random
from matplotlib import pyplot as plt

T = 24*60 
l1 = 1/10

del_t = 0.1
n = 1000
t = []

for i in range(n):
  tk = 0
  t1 = 0
  t3 = random.uniform(0,T)
  while (tk < T):
    tk += del_t


    if (random.uniform(0,1) < l1 * del_t):
      t2 = tk
      if (t3 < tk):
        break
      t1 = tk

  t += [t2-t1]

sum = 0
for i in t:
  sum += i

t0 = sum/n

print("The mean interarrival time noted by Inspector : "+ str(t0))
print("It's twice the real mean interarrival time of buses i.e. "+ str(1/l1))

t_set = set(t)
t_dict = {}

for i in t_set:
  t_dict[i] = 0

for i in t:
  t_dict[i] += 1/n

plt.hist(t , edgecolor = "red",bins = 16)
plt.xlabel("Interarrival Time Noted")
plt.ylabel("Number of Arrivals (out of "+str(n)+" trials)")
plt.show()

bin_size = 2
bin_max = max(t_set)
bins = 0
while (bins < bin_max):
  bins += bin_size

p = {}
for i in range(0,bins+1):
  p[i*bin_size] = 0

for i in p.keys():
  for j in t_dict.keys():
    if (i-bin_size < j < i):
      p[i] += t_dict[j]


x = list(p.keys())
y = list(p.values())

plt.figure(figsize=(20, 5))
plt.bar(x,y,color='blue',edgecolor='red',width=0.8)

plt.xlabel("Interarrival Time (Bin size:"+ str(bin_size)+") ")
plt.ylabel("Probability")
plt.show()





```
