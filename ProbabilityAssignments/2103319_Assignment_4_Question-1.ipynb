```python
import random
from matplotlib import pyplot as plt

p = 0.5
n = 10000

t = []
nt = {}

k = 10


for i in range(n):
  c = 0
  j = 0
  while (c < k):
    j += 1
    if random.uniform(0,1) < p:

      c += 1

  t+=  [j]

set_t = set(t)



dict_t = {}

for i in set_t:
  dict_t[i] = 0
for i in t:
  dict_t[i] += 1


prob_t = {}

for i in dict_t:
 prob_t[i] = dict_t[i]/n
  

print(prob_t)

y = list(prob_t.values())
x = list(prob_t.keys())

plt.bar(x,y,color='blue',edgecolor='red',width=0.5)
plt.xlabel(str(k) +"th success at")
plt.ylabel("Probability")
plt.title("Empirical Distribution")
plt.show()



    
    
```
