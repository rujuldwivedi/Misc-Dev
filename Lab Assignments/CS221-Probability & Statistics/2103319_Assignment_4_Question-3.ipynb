```python
import random
from matplotlib import pyplot as plt

T = 20
nk = 1000
n = 10000
del_t = T/nk
l1 = 1
l2 = 2
l3 = l1 + l2 

k_seq = []

k_comb = []

for i in range(n):
  
  k = 0
  k3 = 0
  for j in range(nk):
    
    if(random.uniform(0,1) < l1*del_t):
      k += 1
    elif (random.uniform(0,1) < l2*del_t):
      k += 1
   

    if (random.uniform(0,1) < l3*del_t): 
      k3 += 1
  
  k_seq += [k3]

  k_comb += [k]

k_set = set(k_comb)
k_dict = {}

for i in k_set:
  k_dict[i] = 0

for i in k_comb:
  k_dict[i] += 1/n


x = list(k_dict.keys())
y = list(k_dict.values())

plt.bar(x,y,color='blue',edgecolor='blue',width=0)
plt.xlabel("Number of arrivals")
plt.ylabel("Probability")
plt.title("Combined Queue Probability Distribution")
plt.show()

seq_set = set(k_seq)
seq_dict = {}

for i in seq_set:
  seq_dict[i] = 0
for i in k_seq :
  seq_dict[i] += 1/n


x = list(seq_dict.keys())
y = list(seq_dict.values())

plt.bar(x,y,color='blue',edgecolor='blue',width=0)
plt.xlabel("Number of arrivals")
plt.ylabel("Probability")
plt.title("Probability Distribution Poisson Process")
plt.show()




```
