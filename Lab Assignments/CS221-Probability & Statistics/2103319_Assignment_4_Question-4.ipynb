```python
import random
from matplotlib import pyplot as plt


l = 2
del_t = 0.0001
n = 10000

t = []

k = 10

for i in range(1000):
  i = 0
  c = 0
  while (i < k):
    if (random.uniform(0,1) < l*del_t):
      i += 1
    c += del_t
  
  t += [c]


plt.hist(t,edgecolor="blue", bins = 20)
plt.xlabel("Time till "+str(k)+ "th arrival")
plt.ylabel("Probability")
plt.title("Empirical Distribution")
plt.show()



```
