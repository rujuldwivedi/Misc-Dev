```python
import random
from matplotlib import pyplot as plt

p1 = 1/4
p2 = 1/4
M = 4
N = 100
n = 10000;

pi = 1/(1+M)

state = []
for i in range(M+1):
  state += [i]



state_n = []

for i in range(n):
  c = 0
  for j in range(N):
    r = random.uniform(0,1)
    
    if (r < p1):
      if (c != M):
        c = c+1

    elif (r < p1+p2):
      if (c != 0):
          c = c-1

  state_n += [c]


set_n = set(state_n)

prob_n = {}

for i in set_n:
  prob_n[i] = 0

for i in state_n:
  prob_n[i] += 1/N

print("Probabilties of final states: ",prob_n)

x = list(prob_n.keys())
y = list(prob_n.values())

plt.bar(x,y,color='blue',edgecolor='blue',width=0.8)
plt.xlabel("Final State at n = "+str(N))
plt.ylabel("Probability")
plt.title("Theoretically π = "+str(pi))
plt.show()
```
